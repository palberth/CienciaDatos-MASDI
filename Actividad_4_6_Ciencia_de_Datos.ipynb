### 3.1 Reemplazo de valores faltantes
Se utiliza el metodo de:
**Imputar los valores faltantes**  
   - **Descripción**: Reemplazar valores faltantes con estimaciones (media, mediana, moda, KNN, regresión).
   - **Cuándo usar**: Para preservar el tamaño de los datos.
   - **Ventajas**: Evita la pérdida de información.
   - **Desventajas**: Puede introducir sesgo.


```python
# Reemplazamos los valores nulos por la media de la columna
data['Esperanza de vida'] = data['Esperanza de vida'].fillna(data['Esperanza de vida'].mean())
data['Mortalidad adulta'] = data['Mortalidad adulta'].fillna(data['Mortalidad adulta'].mean())
data['Consumo de alcohol'] = data['Consumo de alcohol'].fillna(data['Consumo de alcohol'].mean())
data['Vacunación Hepatitis B'] = data['Vacunación Hepatitis B'].fillna(data['Vacunación Hepatitis B'].mean())
data['Índice de masa corporal (IMC)'] = data['Índice de masa corporal (IMC)'].fillna(data['Índice de masa corporal (IMC)'].mean())
data['Vacunación contra la poliomielitis'] = data['Vacunación contra la poliomielitis'].fillna(data['Vacunación contra la poliomielitis'].mean())
data['Gasto total'] = data['Gasto total'].fillna(data['Gasto total'].mean())
data['Vacunación contra la difteria'] = data['Vacunación contra la difteria'].fillna(data['Vacunación contra la difteria'].mean())
data['PIB'] = data['PIB'].fillna(data['PIB'].mean())
data['Población'] = data['Población'].fillna(data['Población'].mean())
data['Delgadez 1-19 años'] = data['Delgadez 1-19 años'].fillna(data['Delgadez 1-19 años'].mean())
data['Delgadez 5-9 años'] = data['Delgadez 5-9 años'].fillna(data['Delgadez 5-9 años'].mean())
data['Composición de ingresos de los recursos'] = data['Composición de ingresos de los recursos'].fillna(data['Composición de ingresos de los recursos'].mean())
data['Escolaridad'] = data['Escolaridad'].fillna(data['Escolaridad'].mean())
```

### 6. Preprocesamiento de datos
Incluye imputación de valores faltantes, normalización y codificación de variables categóricas.


### 6.1 Importación de Librerias
Realizamos la importación de las librerias necesarias

```python
# Importar librerías necesarias
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
import matplotlib.pyplot as plt
```

### 6.2 Implementación de K-Medias

```python

# Escalar los datos usando MinMaxScaler
scaler = MinMaxScaler()
# Seleccione solo características numéricas para escalar
numerical_features = data.select_dtypes(include=['number']).columns
data_scaled = scaler.fit_transform(data[numerical_features])

# Determinar el número óptimo de clústeres con el método del codo
sse = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data_scaled)  # Now data_scaled is defined and can be used
    sse.append(kmeans.inertia_)

# Gráfico del método del codo
plt.plot(range(2, 11), sse, marker='o')
plt.title('Método del codo')
plt.xlabel('Número de clústeres')
plt.ylabel('SSE')
plt.show()

# K-Medias con k óptimo
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans_labels = kmeans.fit_predict(data_scaled)
```

### 6.3 Implementación de métodos aglomerativos



```python
# Aglomerativo
agglo = AgglomerativeClustering(n_clusters=4, linkage="average")
agglo_labels = agglo.fit_predict(data_scaled)
```

DVSCAN

```python
# DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(data_scaled)
```

### 7. Evaluación y visualización de métricas
Implementamos modelos y generamos las métricas.


```python
# Evaluación de clústeres
models = {
    "K-Medias": kmeans_labels,
    "Aglomerativo": agglo_labels,
    "DBSCAN": dbscan_labels
}

for name, labels in models.items():
    silhouette = silhouette_score(data_scaled, labels) if len(set(labels)) > 1 else -1
    calinski = calinski_harabasz_score(data_scaled, labels) if len(set(labels)) > 1 else -1
    davies = davies_bouldin_score(data_scaled, labels) if len(set(labels)) > 1 else -1
    print(f"{name} - Silueta: {silhouette:.2f}, Calinski-Harabasz: {calinski:.2f}, Davies-Bouldin: {davies:.2f}")

# Gráfico de dispersión para visualizar clústeres
plt.scatter(data_scaled[:, 0], data_scaled[:, 1], c=kmeans_labels, cmap='viridis', s=50)
plt.title('Clústeres K-Medias')
plt.show()
```

### 7.1 visualización 3D de Clústeres K-Means
Seleccionando variables relevantes, como PIB, gasto en salud y mortalidad adulta, considerando su correlación con la expectativa de vida

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

df = data
# Preprocesamiento: Limpieza y normalización
df = df.dropna()  # Eliminar filas con valores nulos
variables_relevantes = ['PIB', 'Gasto total', 'Mortalidad adulta', 'Esperanza de vida '] # Lista original

# Imprimir los nombres de las columnas disponibles en el DataFrame
print(f"Columnas disponibles en el DataFrame: {df.columns.tolist()}")

# Corregir los nombres de las columnas en 'variables_relevantes' si es necesario
variables_relevantes = [col.strip() for col in variables_relevantes] # Eliminar espacios al inicio y al final
# Verificar si las columnas existen en el DataFrame y si no, imprimir una advertencia y excluirlas
variables_relevantes = [col for col in variables_relevantes if col in df.columns]

# Asegurarse de que 'variables_relevantes' no esté vacía después de la corrección
if not variables_relevantes:
    raise ValueError("Ninguna de las 'variables_relevantes' se encuentra en el DataFrame. Revise los nombres de las columnas.")

df = df[variables_relevantes] # Seleccionar solo las columnas relevantes
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)

# Aplicar PCA para reducir dimensiones (para visualización 3D)
pca = PCA(n_components=3)
df_pca = pca.fit_transform(df_scaled)

# K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(df_pca)
labels = kmeans.labels_

# Gráfico 3D de los clústeres
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

# Asignar colores a los clústeres
colors = ['red', 'blue', 'green', 'orange']
for i in range(4):
    cluster_points = df_pca[labels == i]
    ax.scatter(cluster_points[:, 0], cluster_points[:, 1], cluster_points[:, 2],
               c=colors[i], label=f'Clúster {i + 1}', s=50)

ax.set_title('Visualización de Clústeres K-Means en 3D')
ax.set_xlabel('Componente Principal 1')
ax.set_ylabel('Componente Principal 2')
ax.set_zlabel('Componente Principal 3')
ax.legend()
plt.show()
```

```python
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

# Calcular métricas
silhouette_avg = silhouette_score(df_pca, labels)
calinski_harabasz = calinski_harabasz_score(df_pca, labels)
davies_bouldin = davies_bouldin_score(df_pca, labels)

print("Métricas de Evaluación:")
print(f"Índice de Silueta: {silhouette_avg:.2f}")
print(f"Índice de Calinski-Harabasz: {calinski_harabasz:.2f}")
print(f"Índice de Davies-Bouldin: {davies_bouldin:.2f}")
```
